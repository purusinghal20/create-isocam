## API.user_timeline([id/user_id/screen_name][, since_id][, max_id][, count][, page])
Returns the 20 most recent statuses posted from the authenticating user or the user specified. It‚Äôs also possible to request another user‚Äôs timeline via the id parameter.

Parameters:	
id ‚Äì Specifies the ID or screen name of the user.

user_id ‚Äì Specifies the ID of the user. Helpful for disambiguating when a valid user ID is also a valid screen name.

screen_name ‚Äì Specifies the screen name of the user. Helpful for disambiguating when a valid screen name is also a user ID.

since_id ‚Äì Returns only statuses with an ID greater than (that is, more recent than) the specified ID.

max_id ‚Äì Returns only statuses with an ID less than (that is, older than) or equal to the specified ID.

count ‚Äì Specifies the number of statuses to retrieve.

page ‚Äì Specifies the page of results to retrieve. Note: there are pagination limits.

```python
name = "BillGates" 
posts = api.user_timeline(screen_name = name, count = 100 , lang = "en" , tweet_mode = "extended")

print("Five tweets from " + name +  "\n")
i = 1
for post in posts[0:5]:
  print(str(i) + ")" + post.full_text + "\n")
  i +=1 
```

```python
df = pd.DataFrame([post.full_text for post in posts] , columns=["Tweets"])
df.head()
```

```python
# Create a function to clean the tweets
def cleanTxt(text):
 text = re.sub('@[A-Za-z0‚Äì9]+', '', text) #Removing @mentions
 text = re.sub('#', '', text) # Removing '#' hash tag
 text = re.sub('RT[\s]+', '', text) # Removing RT
 text = re.sub('https?:\/\/\S+', '', text) # Removing hyperlink
 
 return text


# Clean the tweets
df['Tweets'] = df['Tweets'].apply(cleanTxt)

# Show the cleaned tweets
df
```

subjectivity
  Return the subjectivity score as a float within the range [0.0, 1.0] where 0.0 is very objective and 1.0 is very subjective.

```python
def get_subjectivity(text):
  return TextBlob(text).sentiment.subjectivity
def get_polarity(text):
  return TextBlob(text).sentiment.polarity

df["Subjectivity"] = df["Tweets"].apply(get_subjectivity)
df["Polarity"] = df["Tweets"].apply(get_polarity)
df
```

```python
allwords = " ".join([twt for twt in df['Tweets']])
wordcloud = WordCloud(width=700 , height=500  , margin = 2 , max_font_size=100 , background_color="White" , random_state=30).generate(allwords)
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
plt.show()
```

```python
wordcloud.to_file("first_review1.png")
```

```python
def get_analysis(score):
  if score < 0:
    return "Negative"
  elif score == 0:
    return "Neutral"
  else:
    return "Positive"
df["Analysis"] = df["Polarity"].apply(get_analysis)
df.head()
```
